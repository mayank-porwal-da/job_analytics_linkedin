```python
import pandas as pd # Importing libraries
import numpy as np
```

```python
df = pd.read_csv('master_data.csv')
df.drop(['Unnamed: 0','Job-id','Company-ID','Job_description'],axis=1, inplace=True)
df
```

```python
for col in ['involvement','Applicants','Level']:
    df[col] = df[col].astype('str').apply(lambda x: x.strip())
```

# Details Table:

```python
details_table = df[['involvement','Applicants','Level']]
```

```python
details_table.drop_duplicates(inplace=True)
```

```python
details_table.reset_index(drop=True, inplace=True)
```

```python
details_table.reset_index(inplace=True)
```

```python
details_table.columns=['Details_ID', 'involvement', 'Applicants', 'Level']
```

```python
details_table.Details_ID = details_table.Details_ID.astype('str').apply(lambda x: 'D-'+x )
```

```python
details_table
```

Left joining details_table on the main DF:

```python
new_df= pd.merge(df,details_table, how='left',on=['involvement', 'Applicants', 'Level'])
```

```python
new_df
```

# Company table:

```python
company_table = df[['Company','Industry','Employees','Followers']]
```

```python
company_table.drop_duplicates(inplace=True) # Removing duplicate rows from company_table
company_table.reset_index(drop=True, inplace=True)  
```

```python
company_table
```

```python
company_table.reset_index(inplace=True)
```

```python
company_table
```

```python
company_table.columns=['Company_ID', 'Company', 'Industry', 'Employees', 'Followers']
company_table.Company_ID= company_table.Company_ID.astype('str').apply(lambda x: 'C-'+x)
```

```python
company_table
```

Left Joining company_table on the main data frame:

```python
new_df=pd.merge(new_df,company_table,how='left', on=['Company', 'Industry', 'Employees', 'Followers'])
```

# Job table:

```python
new_df
```

```python
job_table = new_df[['Job_name','Location','Company_ID','Details_ID',]]
```

```python
job_table.drop_duplicates(inplace=True)
job_table.reset_index(drop=True, inplace=True)
```

```python
job_table.reset_index(inplace=True)
```

```python
job_table.columns=['Job_ID', 'Job_name', 'Location', 'Company_ID', 'Details_ID']
job_table.Job_ID = job_table.Job_ID.astype('str').apply(lambda x: 'J-'+x)
```

```python
job_table
```

Left joining job_table on new_df:

```python
new_df = pd.merge(new_df,job_table,how='left', on=['Job_name', 'Location', 'Company_ID', 'Details_ID'])
```

```python
new_df
```

# Converting every DataFrame to csv files

```python
new_df.to_csv('new_df.csv')
company_table.to_csv('company_table.csv')
job_table.to_csv('job_table.csv')
details_table.to_csv('details_table.csv')
```
